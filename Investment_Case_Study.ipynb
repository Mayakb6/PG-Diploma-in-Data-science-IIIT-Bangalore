```python
#Importing the mapping file provided for sector analysis
sector_mapping = pd.read_csv(r"E:\DATA SCIENCE IIIT-B\case study\mapping.csv", sep = ",", encoding = "ISO-8859-1")
sector_mapping
```

```python
#Getting the last 150 rows
sector_mapping.tail(150)
```

```python
#Replacing wrong spellings with NA as data the cleaning because few names have '0' below code will help in correcting the spelling
sector_mapping['category_list'] = sector_mapping['category_list'].str.replace('0', 'na')
```

```python
#Checking for notnull values
sector_mapping= sector_mapping[sector_mapping['category_list'].notnull()]
```

```python
#Drop Rows dor master frame where category_list is null
master_frame= master_frame[~master_frame['category_list'].isnull()]
master_frame.shape
```

```python
##dropping Blanks column as it does not have any needed data
sector_mapping.drop(columns=['Blanks'],inplace =True)
sector_mapping
```

```python
#checking for shape
sector_mapping.shape
```

```python
#creating mapping frame with id and value
final = pd.melt(sector_mapping, id_vars=["category_list"], var_name='category', value_name = 'flag')
```

```python
### appending value 1 to flag
final = final[final['flag'] == int(1)]
final
```

```python
### extracting the primary sector for all the companies 
master_frame["primary_sector"] = master_frame['category_list'].str.split('|', n=1)
master_frame['primary_sector'] = master_frame['primary_sector'].apply(lambda x: x[0])
master_frame
```

```python
## merging the frames final and master_frame as it will main sector of all the companies and converting the flag to int
final_frame = master_frame.merge(final, left_on = 'primary_sector' , right_on = 'category_list' , how = 'left')
```

```python
#### null values in main_sector because category 'Nanotechnology' dont have main sector in final df
final_frame.dropna(axis =0, inplace = True)
final_frame
```

```python
final_frame.drop(columns = 'category_list_y', inplace= True)
final_frame.rename(columns={'category_list_x':'category_list', 'category':'main_sector'}, inplace = True)
final_frame['flag'] = final_frame['flag'].apply(lambda x : int(x))
final_frame
```

```python
## checking for the null values
final_frame.isna().sum()
```

```python
##checking for the number of companies in each main sector
final_frame.groupby('main_sector').sum()['flag'].sort_values(ascending = False)
```
